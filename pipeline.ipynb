#pipeline

最简单完成NLP任务就是直接使用pipeline，也就是官方训练好的模型，设置好的接口，直接使用即可，目前可以pipeline可以完成

nlp

*       sentiment analysis 情感分析
*       text generation   文本生成
*       name entity recognation 命名实体识别
*       Question Anwsering  问答
*       Fill-mask      填空
*       Translation     翻译
*       Feature extraction 特征提取
*       Summerization    摘要

cv

*       Image classification 图像分类
*       Image segmentation  图像分割
*       Object Detection   目标识别

audio

*       audio classification 语音分类
*       automatic speech recognition ASR 语言转文字

```python
#使用pipeline
from transformers import pipeline

classifier=pipeline('sentiment-analysis')

#可以单一文本
text='when something went wrong ,i will go to play basketball or other sports to release myself from the mess.'
r1=classifier(text)

#可以多个文本list
texts=['when something went wrong ,i will go to play basketball or other sports to release myself from the mess.','I love you.']
r2=classifier(texts)

print(r1)
print(r2)
```

除了单一文本和list，还可以对整个数据集dataset进行预测

```python
#这里使用cuda进行加速预测
import datasets
from tqdm.auto import tqdm
from transformers.pipelines.pt_utils import KeyDataset

speech_recognizer=pipeline('automatic-speech-recognition',model='facebook/wav2vec2-base-960h',device=0)

dataset=datasets.load_dataset('superb',name='asr',split='test')

for out in tqdm(speech_recognizer(KeyDataset(dataset,'file'))):
  print(out)
```
